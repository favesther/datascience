# Load and explore the data
1. load the sd-data.csv demo data and get a quick overview of it. (observations, variables, values)
2. Make a k-NN test plot where you predict the feature colors. You can use the following example
code:

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier
#

data = pd.read_csv("2d-data.csv.bz2", sep="\t")
X = data[['X1', 'X2']].values
y = data.Y.values
## image grid
nGrid = 100
range1 = np.linspace(X[:,0].min(), X[:,0].max(), nGrid)
range2 = np.linspace(X[:,1].min(), X[:,1].max(), nGrid)
(x02d, x12d) = np.meshgrid(range1, range2)
x0 = x02d.ravel()
x1 = x12d.ravel()
xp = np.column_stack((x0, x1))
## example prediction w/kNN
mNN = KNeighborsClassifier(n_neighbors=1).fit(X, y)
ypNN = mNN.predict(xp)
ypNN2d = ypNN.reshape((nGrid,nGrid))
plt.imshow(ypNN2d, alpha=0.3, extent=[x0.min(), x0.max(), x1.min(), x1.max()],
origin='lower', aspect='auto')
plt.scatter(X[:,0], X[:,1], c=y, s=100, alpha=0.5)
plt.show()
```
