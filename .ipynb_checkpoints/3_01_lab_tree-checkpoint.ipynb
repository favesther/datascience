# Trees, Bagging, Boosting, Forests

```python
import numpy as np
import pandas as pd
import scipy.special as ss
from sklearn import tree, ensemble
from sklearn.model_selection import train_test_split, cross_validate
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
```

## Create random data

Create random dots that are categorized by color along the spiral arms

```python
n=999
x=np.random.normal(size=n)
y=np.random.normal(size=n)
alpha=np.arctan2(y,x)
r=np.sqrt(x**2 + y**2)
c1 = (np.sin(2*alpha + 2*r) > 0).astype('int')
c2 = (np.cos(2*alpha + 2*r) > 0).astype('int')
c = c1 + c2  # c is color for the spiral arms
```

$\alpha=\arctan(y,x)$  
$r=\sqrt(x^2 + y^2)$   
$c1 = (\sin(2*\alpha + 2*r) > 0)$    
$c2 = (\cos(2*\alpha + 2*r) > 0)$

```python
plt.figure(figsize=(12,12))# size in inches (sort of), find one suitable to you
plt.scatter(x, y, c=c, s=50, edgecolors='k')
X=np.column_stack((x,y))
Xtrain, Xtest, ctrain, ctest = train_test_split(X, c, test_size=0.2)
```

## Plain trees

```python
m = tree.DecisionTreeClassifier(max_depth=None).fit(Xtrain, ctrain)
# max depth=None fits perfectly (and overfits!)
chat=m.predict(Xtrain)
print("accracy on training data:",np.mean(ctrain == chat))
pred = pd.DataFrame({'x': Xtrain[:,0], 'y':Xtrain[:,1], 'C': ctrain,'hatC':chat})
print(pred.sample(5))
chat=m.predict(Xtest)
print("accuracy on testing data: ", np.mean(ctest == chat))
pred = pd.DataFrame({'x':Xtest[:,0],'y': Xtest[:,1],'C':ctest,'hatC':chat})
print(pred.sample(5))
```

```python
nGrid = 100
cmap_light = ListedColormap(['#FFAAAA', '#AAFFAA', '#AAAAFF'])
cmap_bold = ListedColormap(['#FF0000', '#00FF00', '#0000FF'])

x_min,x_max = X[:,0].min() - 1, X[:, 0].max() + 1
y_min,y_max = X[:,1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.linspace(x_min, x_max, nGrid),
                     np.linspace(y_min, y_max, nGrid))
XX = np.column_stack((xx.ravel(),yy.ravel()))
CC = m.predict(XX).reshape(xx.shape)
pred = pd.DataFrame({'x':xx.ravel(),'y': yy.ravel(),'C':CC.ravel()})
print( pred.sample(10))

# Put the result into a color plot
plt.figure(figsize=(9,9))
plt.pcolormesh(xx, yy, CC, cmap=cmap_light)

# Plot also the training points
plt.scatter(Xtest[:,0], Xtest[:,1], c=ctest, s=50, edgecolors='k', cmap=cmap_bold)
plt.xlim(xx.min(), xx.max())
plt.ylim(yy.min(), yy.max())
plt.title("3-Class classification k = %i"
          % (n))
plt.show()
```

# Bagging

```python
m = ensemble.BaggingClassifier(n_estimators=100).fit(Xtrain, ctrain)
# max depth=None fits perfectly (and overfits!)
chat = m.predict(Xtrain)
print("accuracy on training data:", np.mean(ctrain == chat))
pred = pd.DataFrame({'x': Xtrain[:,0],'y':Xtrain[:,1],'C':ctrain,'hatC':chat})
# x, y: coordinates, C: actual color, hatC: predicted color

print(pred.sample(5))
cHat = m.predict(Xtest)
print("accuracy on testing data:", np.mean(ctest == cHat))
pred = pd.DataFrame({'x':Xtest[:,0], 'y':Xtest[:,1], 'C':ctest, 'hatC':cHat})
print(pred.sample(5))
```

```python
nGrid = 100
cmap_light = ListedColormap(['#FFAAAA', '#AAFFAA', '#AAAAFF'])
cmap_bold = ListedColormap(['#FF0000', '#00FF00', '#0000FF'])

x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.linspace(x_min, x_max, nGrid),
                     np.linspace(y_min, y_max, nGrid))
XX = np.column_stack((xx.ravel(), yy.ravel()))
CC = m.predict(XX).reshape(xx.shape)
pred = pd.DataFrame({'x':xx.ravel(), 'y':yy.ravel(), 'C':CC.ravel()})
print(pred.sample(10))

# Put the result into a color plot
plt.figure(figsize=(9,9))
plt.pcolormesh(xx, yy, CC, cmap=cmap_light)

# Plot also the training points
plt.scatter(Xtest[:,0], Xtest[:,1], c=ctest, s=40, edgecolors='k', cmap=cmap_bold)
plt.xlim(xx.min(), xx.max())
plt.ylim(yy.min(), yy.max())
plt.title("3-Class classification k = %i"
          % (n))
plt.show()
```

# random forest

```python
m=ensemble.RandomForestClassifier(n_estimators=100).fit(Xtrain, ctrain)
chat = m.predict(Xtrain)
print("accuracy on training data:", np.mean(ctrain == chat))
pred = pd.DataFrame({'x': Xtrain[:,0],'y':Xtrain[:,1],'C':ctrain,'hatC':chat})
# x, y: coordinates, C: actual color, hatC: predicted color

print(pred.sample(5))
cHat = m.predict(Xtest)
print("accuracy on testing data:", np.mean(ctest == cHat))
pred = pd.DataFrame({'x':Xtest[:,0], 'y':Xtest[:,1], 'C':ctest, 'hatC':cHat})
print(pred.sample(5))
```

# cross-validation

```python
for d in [1,2,3,4,5,10,20,30,40,50,100,None]:
    m = tree.DecisionTreeClassifier(max_depth=d)
    cv = cross_validate(m, X, c, scoring="accuracy", cv=5)
    print("max depth:", d, ", avg test score:", cv['test_score'].mean())
```

max depth 40 is the optimum.
